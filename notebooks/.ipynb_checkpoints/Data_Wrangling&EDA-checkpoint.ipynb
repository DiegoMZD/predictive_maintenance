## Look for correlation between the variables

```python
plt.subplots(figsize=(5,4))
sns.heatmap(data.select_dtypes(["int","float"]).corr());
```

```python
fig, axes = plt.subplots(1, 2, figsize = (12, 5))

cond_fail = data["Target"] == 1
corr1 = data[cond_fail].select_dtypes(["int","float"]).drop(columns="Target").corr()
sns.heatmap(corr1, ax=axes[0])
axes[0].set_title("With Failure")

cond_fail = data["Target"] == 0
corr0 = data[cond_fail].select_dtypes(["int","float"]).drop(columns="Target").corr()
sns.heatmap(corr0, ax=axes[1])
axes[1].set_title("With No Failure")

plt.tight_layout()
plt.show()
```

```python
diff = corr1 - corr0
plt.subplots(figsize=(5,4))
sns.heatmap(diff)
plt.title("Correlation Variation due to the target");
```

```python
sns.scatterplot(data = data, x = "Air temperature [K]", y = "Process temperature [K]", sizes = .5, hue = "Target");
```

```python
sns.scatterplot(data = data, x = "Torque [Nm]", y = "Rotational speed [rpm]", sizes = .5, hue = "Target");
```

```python
data1 = data.copy()
data1["Target"] = data1["Target"].astype(str)
sns.countplot(data = data1, x = "Type", hue = "Target");
```

```python
count_values = data1.groupby(["Type", "Target"]).size()
count_values
```

```python
for type in ["L", "M", "H"]:
    percentage = count_values.loc[(type, '1')] / count_values.loc[(type, '0')] * 100
    print(f"The proportion between fail and not fail, for the type {type} is: {round(percentage, 2)}%")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, type_value in enumerate(["L", "M", "H"]):
    cond_type = data["Type"] == type_value
    type_data = data[cond_type].select_dtypes(["int", "float"]).drop(columns="Target")
    
    corr_matrix = type_data.corr()
    
    sns.heatmap(corr_matrix, ax=axes[i])
    axes[i].set_title(f"Type {type_value}")

plt.tight_layout()
plt.show()
```
